```python
import csv
import time
import pandas as pd
import re
import json
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep

chromedriver = r'C:\Users\bjh99\Downloads\chromedriver_win32 (1)\chromedriver.exe'
wd = webdriver.Chrome(chromedriver)
```

```python
wd.get("https://m.place.naver.com/restaurant/723048338/review/visitor")
```

```python
arr_review = []
arr_title = []
req = wd.page_source
soup = BeautifulSoup(req,'html.parser')
review = soup.find_all('a',"xHaT3")
title = soup.find('span','Fc1rA').text
for i in range(len(review)):
    if len(review[i]) == 0:
        continue
    
    pattern=re.compile("[ㄱ-ㅎㅏ-ㅣ가-힣]+")
    result=pattern.findall(str(review[i]))
    if len(result)==0:
        continue
    result = " ".join(result)
    arr_review.append(result)
    arr_title.append(title)

total = pd.DataFrame({'리뷰':arr_review, '타이틀':arr_title})
total.to_csv('차차돈까스.csv',index=False)
print(len(arr_review))
```

```python
import re
ii = []
data_1=TOTAL.dropna(axis=0)
arr_review=[]
arr_review_1=[]
for i in range(len(data_1)):
    arr_review.append(data_1.iloc[i,0])
    arr_review_1.append(data_1.iloc[i,0])
from konlpy.tag import Okt
okt=Okt()

for i in range(len(arr_review)):
    arr_review[i]=re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]"," ", arr_review[i])
    arr_review[i]=okt.morphs(arr_review[i],norm=True, stem=True)
    arr_review[i]=[x for x in arr_review[i] if len(x)>1]
    arr_review[i]=' '.join(arr_review[i])
data_2=pd.DataFrame({'review':arr_review,'raw':arr_review_1})
hangul = re.compile('[ㄱ-ㅣ가-힣]+')
for i in range(len(data_2)):
    if len(hangul.findall(data_2.iloc[i,0]))>1:
        if len(data_2.iloc[i,0])>1:
            ii.append(i)

data_3=data_2.iloc[ii,:]
data_3.to_csv('test.csv',index=False)
```

```python
A=data_3.review.to_list()
B=[]
for i in range(len(A)):
    B+=A[i].split(' ')
```

```python
words_count={}
for word in B:
    if word in words_count:
        words_count[word] += 1
    else:
        
        words_count[word] = 1
sorted_words = sorted([(k,v) for k,v in words_count.items()], key=lambda word_count: -word_count[1])
print(sorted_words)
#print([w[0] for w in sorted_words])
```

```python
new_words = []
for word in sorted_words:
    if word[1]>9:
        new_words.append(word)
print(len(new_words))
```

```python
from gensim.models import Word2Vec
```

```python
X=[[]]*len(data_3)
for i in range(len(data_3)):
    X[i] = data_3.iloc[i,0].split(' ')
```

```python
model_w = Word2Vec(sentences=X,window=5, min_count=10, workers=6,sg=1)
```

```python
df1=pd.DataFrame(model_w.wv.most_similar('맛있다',topn=2000),columns=['단어','맛있다'])
```
